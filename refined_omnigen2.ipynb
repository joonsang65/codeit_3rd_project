# 필요한 라이브러리 및 프레임워크 불러오기

```python
import accelerate, os, requests, sys, time, torch
from pathlib import Path
from google.colab import userdata
from googlesearch import search
from bs4 import BeautifulSoup # For searching and parsing web content.
from typing import List, Tuple # Pydantic.
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
from torchvision.transforms.functional import to_tensor
from openai import OpenAI

root_dir = Path().resolve()

# Convert PosixPath to string before appending.
sys.path.append(str(root_dir))

# Local directories.
from omnigen2.pipelines.omnigen2.pipeline_omnigen2 import OmniGen2Pipeline
from omnigen2.models.transformers.transformer_omnigen2 import OmniGen2Transformer2DModel
from omnigen2.utils.img_util import create_collage
from omnigen2.pipelines.omnigen2.pipeline_omnigen2_chat import OmniGen2ChatPipeline

# Hugging Face token and OpenAI API key.
HF_TOKEN = userdata.get("HF_TOKEN")
OPENAI_KEY = userdata.get("OPENAI_KEY")

llm = OpenAI(api_key=OPENAI_KEY)
```

# 한국어 프롬프트 영어 프롬프트로 번역

```python
# 한국어 입력 프롬프트.
korean_prompt = input("프롬프트 입력하길 바랍니다: ")

# 시간 측정.
start_time_total = time.time()
# 번역 시간.
start_time_translation = time.time()
```

```python
try:
    translator_response = llm.chat.completions.create(
        model='gpt-4o-mini',
        messages=[
            {
                'role': 'system',
                'content': '''
                You are a Korean-to-English translator who will be translating a user-inputted prompt in Korean about an advertisement-creating request such as how the user wants the background advertisement image generated,
                under what lighting, the ambiant, the mood, and the setting. Ensure that the user-inputted query is translated as accurately as possible into English and avoid typos at all costs.
                '''
            },
            {
                'role': 'user',
                'content': korean_prompt
            }
        ]
    )
    translated_prompt = translator_response.choices[0].message.content
    print(translated_prompt)

except Exception as e:
    print(f"An error occurred during translation: {e}")
    translated_prompt = None # Set to None or handle appropriately

end_time_translation = time.time()
translation_duration = end_time_translation - start_time_translation
print(f"번역 시간 소요: {translation_duration:.2f} 초.")
```

# 번역된 프롬프트 바탕으로 프롬프트 재구성

```python
# 프롬프트 생성 시간 측정.
start_time_prompt_gen = time.time()
prompt_generator = llm.chat.completions.create(
    model='gpt-4o-mini',
    messages=[
        {
            'role': 'system',
            'content': '''
            You are a prompt generator for image creation.
            Translate the user's request for an advertisement image background, originally translated from Korean, into a detailed English prompt.
            The prompt should describe the scene including setting, lighting, mood, objects, theme, angle, and art style.
            Follow this template: [Subject], [Environment], [Style/Lighting], [Camera/Details], [Artist or Art Style].
            If art style, lighting, or camera details are not specified, use clear lighting, photo-realistic style, and shallow depth of field.
            '''
        },
        {
            'role': 'user',
            'content': translator_response.choices[0].message.content
        }
    ]
)

end_time_prompt_gen = time.time()
prompt_gen_duration = end_time_prompt_gen - start_time_prompt_gen
print(f"첫 프롬프트 생성 시간 소요: {prompt_gen_duration:.2f} 초.")

generated_prompt = prompt_generator.choices[0].message.content
print(generated_prompt)
```

번역된 프롬프트에서 웹 검색에 사용할 주요 주제나 키워드를 추출하기 위해 gpt-4o-mini 모델을 사용한다.

```python
# Keyword
# 키워드 검색 시간 측정.
start_time_keyword_extraction = time.time()
keyword_extractor = llm.chat.completions.create(
    model='gpt-4o-mini',
    messages=[
        {
            'role': 'system',
            'content': '''
            You are a keyword extractor for image generation prompts.
            Analyze the user's prompt, which is a translated request for an advertisement image background, and identify the main subject or key keywords that would be most useful for a web search to gather specific details about the subject.
            The subject could be a location (like a beach, city, or landmark), a type of food or drink, or an object.
            Extract only the most relevant terms and present them as a comma-separated list.
            '''
        },
        {
            'role': 'user',
            'content': generated_prompt
        }
    ]
)

end_time_keyword_extraction = time.time()
keyword_extraction_duration = end_time_keyword_extraction - start_time_keyword_extraction
print(f"키워드 검색 시간 소요: {keyword_extraction_duration:.2f} 초.")

extracted_keywords = keyword_extractor.choices[0].message.content
print("Extracted Keywords:", extracted_keywords)
```

# 웹 검색

```python
def perform_web_search(query: str, num_results: int = 5) -> List[str]:
    """Performs a web search and returns a list of result URLs."""
    results = []
    for url in search(query, num=num_results):
        results.append(url)
    return results

# 웹 페이지 검색 시간 측정.
start_time_web_search = time.time()
keyword_search_results = perform_web_search(extracted_keywords)

end_time_web_search = time.time()
web_search_duration = end_time_web_search - start_time_web_search
print(f"웹 페이지 검색 시간 소요: {web_search_duration:.2f} 초.")
print(keyword_search_results)
```

검색 결과를 순회하며 각 URL의 내용을 가져오고, HTML을 파싱한 후 주요 텍스트 내용을 추출하여 리스트에 저장한다.

This takes too long, around 29 minutes.

```python
keyword_text_content = []
# 각 URL 접속해서 텍스트 내용 추출 시간 측정.
start_time_fetch_process = time.time()
for url in keyword_search_results:
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()  # Raise an HTTPError for bad responses (4xx or 5xx)
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract text from common tags that might contain main content
        text = ' '.join([p.get_text() for p in soup.find_all(['p', 'div'])])
        # Truncate the text content to a manageable size
        keyword_text_content.append(text[:2000]) # Adjust truncation length as needed

    except requests.exceptions.RequestException as e:
        print(f"Could not retrieve content from {url}: {e}")
    except Exception as e:
        print(f"An error occurred while processing {url}: {e}")

end_time_fetch_process = time.time()
fetch_process_duration = end_time_fetch_process - start_time_fetch_process
print(f"각 URL 접속해서 텍스트 내용 추출 시간 소요: {fetch_process_duration:.2f} 초.")
print(keyword_text_content)
```

가져온 텍스트 내용에서 사용자 입력과 관련된 정보를 추출하기 위해 LLM을 사용한다.

```python
prompt_refiner = llm.chat.completions.create(
    model='gpt-4o-mini',
    messages=[
        {
            'role': 'system',
            'content': f'''
            You are a prompt refiner for image generation. You will be given an initial prompt translated from Korean and a list of web search results about the subject of the prompt.
            Your task is to refine the initial prompt by incorporating relevant details from the web search results to make the image generation prompt more accurate and detailed.
            Pay close attention to details about the location, such as the presence or absence of specific elements like palm trees.
            Moreover, pay close attention to the presence or absence of specific elements mentioned in the web search results.
            The refined prompt should follow the template: [Subject], [Environment], [Style/Lighting], [Camera/Details], [Artist or Art Style].
            Initial prompt: {generated_prompt}
            Web search results: {keyword_search_results}
            '''
        },
        {
            'role': 'user',
            'content': "Refine the initial prompt based on the web search results."
        }
    ]
)
```

```python
prompt_refiner.choices[0].message.content
```

```python
limited_keyword_text_content = keyword_text_content[:5] # Adjust the number as needed

extracted_keyword_info = llm.chat.completions.create(
    model='gpt-4o-mini',
    messages=[
        {
            'role': 'system',
            'content': f'''
            You are an information extractor for image generation prompts.
            You will receive raw text content from web searches about a subject.
            Your task is to identify and extract key details relevant to generating an image of that subject.
            Focus on aspects such as landmarks, atmosphere, typical activities, and the presence or absence of specific natural elements.
            Present the extracted information in a clear and concise manner.
            The subject is: {extracted_keywords}
            '''
        },
        {
            'role': 'user',
            'content': f"Extract relevant information about {extracted_keywords} from the following text content: {limited_keyword_text_content}"
        }
    ]
)

extracted_keyword_info = extracted_keyword_info.choices[0].message.content
print("Extracted Relevant Information:")
print(extracted_keyword_info)
```

웹 검색 결과에서 추출한 정보를 활용하여 이미지 생성 모델을 위한 더 정확하고 상세한 프롬프트를 생성한다.

```python
# Refine the prompt using extracted information

# 생성한 프롬프트 웹 페이지 검색한 정보 바탕으로 개선 시간 측정.
start_time_prompt_refinement = time.time()
prompt_refiner_with_content = llm.chat.completions.create(
    model='gpt-4o-mini',
    messages=[
        {
            'role': 'system',
            'content': f'''
            You are a prompt refiner for image generation. You will be given an initial prompt translated from Korean and extracted relevant information about the subject from web searches.
            Your task is to refine the initial prompt by incorporating the extracted details to make the image generation prompt more accurate and detailed.
            Pay close attention to the presence or absence of specific elements mentioned in the extracted information.
            The refined prompt should follow the template: [Subject], [Environment], [Style/Lighting], [Camera/Details], [Artist or Art Style].
            Initial prompt: {generated_prompt}
            Extracted information about Haeundae Beach: {extracted_keyword_info}
            '''
        },
        {
            'role': 'user',
            'content': "Refine the initial prompt based on the extracted information."
        }
    ]
)

end_time_prompt_refinement = time.time()
prompt_refinement_duration = end_time_prompt_refinement - start_time_prompt_refinement
print(f"생성한 프롬프트 웹 페이지 검색한 정보 바탕으로 개선 시간 소요: {prompt_refinement_duration:.2f} 초.")

refined_prompt_with_content = prompt_refiner_with_content.choices[0].message.content
print("Refined Prompt with Content:")
print(refined_prompt_with_content)
```

일반적인 부정 프롬프트를 정의하고, 최종 프롬프트를 기반으로 LLM을 사용하여 동적 부정 프롬프트 항목을 생성한 후 결합한다.

```python
# Define a general negative prompt
general_negative_prompt = "(((deformed))), blurry, over saturation, bad anatomy, disfigured, poorly drawn face, mutation, mutated, (extra_limb), (ugly), (poorly drawn hands), fused fingers, messy drawing, broken legs censor, censored, censor_bar"

# 부정 프롬프트 개선 시간 측정.
start_time_neg_prompt_gen = time.time()

# Use LLM to generate dynamic negative prompt terms based on the translated prompt
dynamic_negative_prompt_generator = llm.chat.completions.create(
    model='gpt-4o-mini',
    messages=[
        {
            'role': 'system',
            'content': '''
            You are a negative prompt generator for image creation.
            Based on the following positive prompt (a request for an advertisement image background), identify potential undesirable elements or qualities specific to the subject or scene that should be avoided in the generated image.
            Provide these as a comma-separated list of terms or phrases to be used in a negative prompt.
            Focus on aspects that could detract from a positive advertisement image, such as unnatural features, incorrect details for the subject, or poor visual representation.
            '''
        },
        {
            'role': 'user',
            'content': f"Positive prompt: {refined_prompt_with_content}"
        }
    ]
)

end_time_neg_prompt_gen = time.time()
neg_prompt_gen_duration = end_time_neg_prompt_gen - start_time_neg_prompt_gen
print(f"부정 프롬프트 시간 소요: {neg_prompt_gen_duration:.2f} 초.")

dynamic_negative_terms = dynamic_negative_prompt_generator.choices[0].message.content

# Combine the general and dynamic negative prompts
combined_negative_prompt = general_negative_prompt + ", " + dynamic_negative_terms
print("Combined Negative Prompt:")
print(combined_negative_prompt)
```

# 파이프라인 초기화.

```python
# 모델 불러오는 시간 측정.
start_time_model_load = time.time()
accelerator = accelerate.Accelerator()

model_path="OmniGen2/OmniGen2"
pipeline = OmniGen2Pipeline.from_pretrained(
    model_path,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    token=HF_TOKEN,
  )
pipeline.transformer = OmniGen2Transformer2DModel.from_pretrained(
    model_path,
    subfolder="transformer",
    torch_dtype=torch.bfloat16,
)
pipeline = pipeline.to(accelerator.device, dtype=torch.bfloat16)

end_time_model_load = time.time()
model_load_duration = end_time_model_load - start_time_model_load
print(f"모델 불러오는 시간 소요: {model_load_duration:.2f} 초.")
```

# 이미지 생성

```python
instructions = [refined_prompt_with_content]

# 생성 이미지 시간 측정.
start_time_image_gen = time.time()
for instruction in instructions:
    generator = torch.Generator(device=accelerator.device).manual_seed(0)
    results = pipeline(
        prompt=instruction,
        input_images=[],
        width=INSTAGRAM[0],
        height=INSTAGRAM[1],
        num_inference_steps=50,
        max_sequence_length=1024,
        text_guidance_scale=4.0,
        image_guidance_scale=1.0,
        negative_prompt=combined_negative_prompt,
        num_images_per_prompt=1,
        generator=generator,
        output_type="pil",
    )

    end_time_image_gen = time.time()
    image_gen_duration = end_time_image_gen - start_time_image_gen
    print(f"이미지 생성 시간 소요: {image_gen_duration:.2f} 초.")

    vis_images = [to_tensor(image) * 2 - 1 for image in results.images]
    output_image = create_collage(vis_images)

    display(output_image)
```

# 광고 문구 생성

```python
# 시각적 이해 시간 측정.
start_time_visual_understanding = time.time()
chat_pipeline = OmniGen2ChatPipeline.from_pipe(pipeline=pipeline, transformer=pipeline.transformer)

# Use the generated image from the previous step
# Assuming 'output_image' holds the PIL Image of the generated advertisement background
# If the variable name is different, please adjust accordingly.
image_for_description = [output_image] # The chat pipeline expects a list of images

# Instruction to get a description suitable for advertisement copy
description_instruction_en = "Please provide a detailed description of this image, focusing on elements, mood, and style that would be useful for writing advertisement copy."

generator = torch.Generator(device=accelerator.device).manual_seed(0)

results = chat_pipeline(
    prompt=description_instruction_en,
    input_images=image_for_description,
    generator=generator,
)

end_time_visual_understanding = time.time()
visual_understanding_duration = end_time_visual_understanding - start_time_visual_understanding
print(f"시각적 이해 시간 소요: {visual_understanding_duration:.2f} 초.")

english_image_description = results.text

print("English Image Description:")
print(english_image_description)
```

### 광고 문구 생성

처음으로 입력한 한국어 프롬프트와 방금 한국어로 번역한 이미지 요약을 바탕으로 광고 문구 생성한다.

```python
# 광고 문구 생성 시간 측정.
start_time_copy_gen = time.time()
ad_copy_generator = llm.chat.completions.create(
    model='gpt-4o-mini',
    messages=[
        {
            'role': 'system',
            'content': f'''
            You are an advertisement copywriter.
            Your task is to generate compelling and creative advertisement copy in Korean for a small business.
            Use the original advertisement request and the description of the generated image to create the copy.
            Ensure the copy is engaging and encourages the target audience to take action (e.g., visit a place, try a product).
            Original Korean Prompt: {korean_prompt}
            Korean Image Description: {korean_image_description}
            '''
        },
        {
            'role': 'user',
            'content': "Generate advertisement copy in Korean based on the provided information."
        }
    ]
)

end_time_copy_gen = time.time()
copy_gen_duration = end_time_copy_gen - start_time_copy_gen
print(f"광고 문구 생성 시간 소요: {copy_gen_duration:.2f} 초.")

korean_advertisement_copy = ad_copy_generator.choices[0].message.content
print("Korean Advertisement Copy:")
print(korean_advertisement_copy)

# --- Calculate Total Time ---
end_time_total = time.time()
total_duration = end_time_total - start_time_total
print(f"\n전 과정 시간 소요: {total_duration:.2f} 초.")
```
